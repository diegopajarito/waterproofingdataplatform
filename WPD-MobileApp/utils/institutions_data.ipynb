````python
df.tipo.replace({'OTHER':"O", 'SCHOOL':"E", 'CIVIL_DEFENSE':"D", 'NGO':"N"}, inplace=True)
````

````python
df.tipo.unique()
````

### Generating the JSON containing the states - cities - institutions

the format is:

```json

```

````python
schools = dict()

for name, group in df.groupby(["uf", "tipo"]):
    uf, tipo = name
    institutions = df.loc[(df.uf == uf) & (df.tipo == tipo)]
    
    if uf not in schools:
        schools[uf] = dict()
    if tipo not in schools[uf]:
        schools[uf][tipo] = list()
        
    [schools[uf][tipo].append({"value":institution,"label":institution}) for institution in institutions.nome]
    

js_content = json.dumps(schools, indent=2, ensure_ascii=False)
````

````python
final_file = "export default " + js_content + ";\n\n"
print(final_file)
````
